# Problem 1 
Kullanıcıdan aldığı ürün bilgisini (ad, fiyat) urunler.txt dosyasına kayıt eden fonksiyonu yazınız.

````python
def urunEkle():

    file = open("urunBilgisi.txt", "r+", encoding="utf-8")
    urun = str(input("Ürün adı giriniz: "))
    fiyat = str(input("Ürün fiyatı giriniz: "))

    with open("urunBilgisi.txt", "r+", encoding="utf-8") as file:
        file.write(f"{urun}, {fiyat}\n")

urunEkle()
````

# Problem 2
1.problemde oluşturulan text dosyası içerisinde kullanıcıdan alacağımız eski kelime isimli değişkene atadığımız 
ve değiştirilmesini istediğimiz kelimemizi yerini alıcak olan yeni kelime isimli değişkene atayın. Belirlenen kelimeleri text dosyasında değiştiren fonksiyonu yazınız.

````python
def urunDegistir():
    eskiUrunAdi = input("Değiştirilecek ürün adını girin: ")
    yeniUrunAdi = input("Yeni ürün adını girin: ")

    with open("urunBilgisi.txt", "r", encoding="utf-8") as file:
        icerik = file.readlines()

    yeniIcerik = [satir.replace(eskiUrunAdi, yeniUrunAdi) for satir in icerik]

    with open("urunBilgisi.txt", "w", encoding="utf-8") as file:
        file.writelines(yeniIcerik)

    print("Kelime değiştirildi.")

urunDegistir()
````

# Problem 3
Vehicle sınıfının tüm değişkenlerini ve yöntemlerini devralacak bir alt sınıf olarak Bus sınıfı oluşturunuz.



```
Output
> Name: School Volvo Speed: 180 Mileage: 12
```


````python
class vehicle():
    def __init__(self, name, speed, mileage):
        self.name = name
        self.speed = speed
        self.mileage = mileage
        print("Araç kaydedildi.")
````

````python
v1 = vehicle("Volvo", "180", "12")
````

````python
class bus(vehicle):
    def __init__(self, name, speed, mileage):
        super().__init__(name, speed, mileage)
        print("Bus kaydedildi.")
    def intro(self):
         print(f"{self.name} markalı otobüsün hızı {self.speed}")
````

````python
b1 = bus("Mercedes", "150", "10")
b1.intro()
````
