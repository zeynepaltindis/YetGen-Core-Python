# Problem 1

Kumanda isminde bir sınıf oluşturarak aşağıdaki işlemleri yapmasını sağlayan fonksiyonları sınıf içerisinde yazalım.
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri


Kumandayı kapatmak için q tuşuna basılmalı ve diğer durumlarda hangi işlemi yapmak istediği sorulmalıdır. Her işlemde gerekli fonksiyonunun sınıf içerisinden çağırılması gerekmektedir.

random ve time kütüphanesini araştırarak kullanabilirsiniz.

Kendi istediğiniz özellikleri eklemekte serbetsiniz.

```python
import random
import time

class Kumanda:
    def __init__(self, tv_durum='Kapalı', tv_ses=0, kanal_listesi=['TRT'], aktif_kanal='TRT'):
        self.tv_durum = tv_durum
        self.tv_ses = tv_ses
        self.kanal_listesi = kanal_listesi
        self.aktif_kanal = aktif_kanal

    def tv_ac(self):
        if self.tv_durum == 'Açık':
            print("Televizyon zaten açık.")
        else:
            print("Televizyon açılıyor...")
            time.sleep(1)
            self.tv_durum = 'Açık'
            print("Televizyon açık.")

    def tv_kapat(self):
        if self.tv_durum == 'Kapalı':
            print("Televizyon zaten kapalı.")
        else:
            print("Televizyon kapanıyor...")
            time.sleep(1)
            self.tv_durum = 'Kapalı'
            print("Televizyon kapalı.")

    def ses_ayarla(self, ses):
        if 0 <= ses <= 100:
            self.tv_ses = ses
            print(f"Ses seviyesi: {self.tv_ses}")
        else:
            print("Ses seviyesi 0 ile 100 arasında olmalıdır.")

    def kanal_ekle(self, kanal):
        print(f"{kanal} kanalı ekleniyor...")
        time.sleep(1)
        self.kanal_listesi.append(kanal)
        print(f"{kanal} kanalı eklendi.")

    def kanal_sayisi_ogren(self):
        return len(self.kanal_listesi)

    def rastgele_kanal_gec(self):
        rastgele_kanal = random.choice(self.kanal_listesi)
        self.aktif_kanal = rastgele_kanal
        print(f"Rastgele kanala geçildi: {self.aktif_kanal}")

    def televizyon_bilgileri(self):
        print(f"TV Durumu: {self.tv_durum}\nSes Seviyesi: {self.tv_ses}\nKanal Listesi: {self.kanal_listesi}\nAktif Kanal: {self.aktif_kanal}")

def main():
    kumanda = Kumanda()

    print("""
    Televizyon Kumandası Uygulaması

    1. TV Aç
    2. TV Kapat
    3. Ses Ayarları
    4. Kanal Ekle
    5. Kanal Sayısını Öğrenme
    6. Rastgele Kanala Geçme
    7. Televizyon Bilgileri
    Çıkmak için 'q' ya basın.
    """)

    while True:
        islem = input("Yapmak istediğiniz işlem: ")

        if islem == 'q':
            print("Programdan çıkılıyor...")
            break
        elif islem == '1':
            kumanda.tv_ac()
        elif islem == '2':
            kumanda.tv_kapat()
        elif islem == '3':
            ses = int(input("Ses seviyesi (0-100): "))
            kumanda.ses_ayarla(ses)
        elif islem == '4':
            kanal = input("Eklenecek kanal ismi: ")
            kumanda.kanal_ekle(kanal)
        elif islem == '5':
            print(f"Kanal sayısı: {kumanda.kanal_sayisi_ogren()}")
        elif islem == '6':
            kumanda.rastgele_kanal_gec()
        elif islem == '7':
            kumanda.televizyon_bilgileri()
        else:
            print("Geçersiz işlem!")

if __name__ == "__main__":
    main()
```
