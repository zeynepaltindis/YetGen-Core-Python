# Problem 1

Şimdi beraber kullanıcılardan aldığımız verileri işleyerek, onlara hem çıktı verecek, hem de aldığımız ve hesapladığımız verileri ileride kullanabilmek adına kaydedeceğiz.


 **Bu bir veri giriş platformu olacak**


 * kullanıcıya kaç giriş yapacağını soracağız
* kullanıcıdan isim-soyisim, yaş, boy, kilo bilgilerini alarak VKİ hesaplayacağız
* ardından bunları sınıflara göre kaydedeceğiz bir dosyaya kaydedeceğiz


Sınıflar ise şu şekilde: 
* 18, 5 kg/m.'nin altında olanlar: Zayıf
* 18.5 – 24, 9 kg/m. arasında olanlar: Normal kilolu
* 25 – 29, 9 kg/m. arasında olanlar: Fazla kilolu
* 30 – 39, 9 kg/m. arasında olanlar: Obez
* 40 kg/m.'nin üzerinde olanlar: İleri derecede obez (morbid obez), olarak görülür.

```python
def vki_hesapla(boy, kilo):
    return kilo / (boy / 100) ** 2

def vki_siniflandir(vki):
    if vki < 18.5:
        return "Zayıf"
    elif 18.5 <= vki < 25:
        return "Normal kilolu"
    elif 25 <= vki < 30:
        return "Fazla kilolu"
    elif 30 <= vki < 40:
        return "Obez"
    else:
        return "İleri derecede obez (morbid obez)"

def veri_girisi():
    kac_giris = int(input("Kaç giriş yapmak istiyorsunuz? "))
    
    veriler = []

    for _ in range(kac_giris):
        isim = input("İsim: ")
        soyisim = input("Soyisim: ")
        yas = int(input("Yaş: "))
        boy = float(input("Boy (cm): "))
        kilo = float(input("Kilo (kg): "))

        vki = vki_hesapla(boy, kilo)
        sinif = vki_siniflandir(vki)

        veri = {
            "isim": isim,
            "soyisim": soyisim,
            "yas": yas,
            "boy": boy,
            "kilo": kilo,
            "vki": vki,
            "sinif": sinif
        }

        veriler.append(veri)

    return veriler

def verileri_dosyaya_kaydet(veriler, dosya_adi="vki_verileri.txt"):
    with open(dosya_adi, 'w') as dosya:
        for veri in veriler:
            dosya.write(f"{veri['isim']} {veri['soyisim']}, Yaş: {veri['yas']}, Boy: {veri['boy']} cm, Kilo: {veri['kilo']} kg, VKİ: {veri['vki']:.2f}, Sınıf: {veri['sinif']}\n")

def main():
    veriler = veri_girisi()
    verileri_dosyaya_kaydet(veriler)
    print("Veriler başarıyla kaydedildi.")

if __name__ == "__main__":
    main()
```

# Problem 2
**YetBar - Smoothie Bank**
---
YetBar spor salonunun girişinde bulunan, insansız bir sağlıklı içecek otomatıdır. Kullanıcı çeşitli girdiler smoothie içeceğine koyacağı ürünleri belirleyebilir, fiyat hesaplaması yaptırabilir ya da sadece menüyü görüntüleyebilir.

**Projenin Amacı:**

*   Birbirini etkileyen ve bir kod içerisinde birden fazla defa başvurulacak fonksiyonların oluşturulup, "fonksiyonel" proglamlamak.
*   Büyük programları yazarken, yazılımcı olarak yapabileceğimiz hatalardan kaçınmak için kontrol yöntemleri geliştirmek.
*   Kullanıcı inputlarını ve yaptığımız hesapları, raporlamak/loglamak amacıyla kaydetmek.

**Programdan Beklenen Çıktılar**
* Program kullanıcıya temelde 2 farklı işlem yapma imkanı sağlar.
> 1. Menüyü görüntüleme
> 3. Satın alma
* Satın alınan her içecek tarih, fiyat bilgisi ile kaydedilmeli ve her zaman ulaşılabilir olmalıdır.




```python
from datetime import datetime

# Basit smoothie menüsü ve fiyatları
menu = {
    "Elma": 5,
    "Muz": 6,
    "Çilek": 7,
    "Kavun": 8
}

def menuGoster(menu):
    print("Menü:")
    for urun, fiyat in menu.items():
        print(f"  {urun}: {fiyat} TL")

def smoothie_hesapla(secimler, menu):
    toplam_fiyat = 0
    for urun in secimler:
        if urun in menu:
            toplam_fiyat += menu[urun]
    return toplam_fiyat

def satin_alma(menu):
    print("Lütfen smoothie için ürünleri seçin (Çıkmak için 'bitir' yazın):")
    secimler = []
    
    while True:
        secim = input("Ürün seçin: ")
        if secim.lower() == 'bitir':
            break
        elif secim in menu:
            secimler.append(secim)
        else:
            print(f"Geçersiz seçim: {secim}")
    
    toplam_fiyat = smoothie_hesapla(secimler, menu)
    islem_kaydet(secimler, toplam_fiyat)
    print(f"Toplam Fiyat: {toplam_fiyat} TL")

def islem_kaydet(secimler, toplam_fiyat):
    with open("satin_alma_kayitlari.txt", "a") as f:
        f.write(f"{datetime.now().strftime('%Y-%m-%d %H:%M:%S')} - Secimler: {', '.join(secimler)} - Toplam Fiyat: {toplam_fiyat} TL\n")

def satin_alma_kayitlari_goster():
    try:
        with open("satin_alma_kayitlari.txt", "r") as f:
            print(f.read())
    except FileNotFoundError:
        print("Henüz satın alma işlemi yapılmamış.")

def main():
    while True:
        print("\nYetBar - Smoothie Bank")
        print("1. Menüyü Görüntüle")
        print("2. Satın Alma")
        print("3. Satın Alma Kayıtlarını Göster")
        print("4. Çıkış")
        
        secim = input("Lütfen bir seçim yapın: ")
        
        if secim == "1":
            menuGoster(menu)
        elif secim == "2":
            satin_alma(menu)
        elif secim == "3":
            satin_alma_kayitlari_goster()
        elif secim == "4":
            print("Çıkış yapılıyor...")
            break
        else:
            print("Geçersiz seçim, lütfen tekrar deneyin.")

if __name__ == "__main__":
    main()
```
